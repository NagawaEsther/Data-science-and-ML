# CREATING ARRAYS

```python
KFC=([7.5,5,4.4,3.75,2.25])
```

```python
CJ_array=np.array(CJ)
CJ_array
```

```python
KFC_array=np.array(KFC)
KFC_array
```

```python
(CJ_array-KFC_array)
```
