```python
import pandas as pd
```

```python
from scipy.stats import linregress
```

```python
# Assuming "df" is your dataframe with"Weight","Height" and columns

#Convert "Weight","Height" and columns to numeric (float)

# Convert "Weight" column to numeric values
df["Weight"] = pd.to_numeric(df["Weight"], errors="coerce")

# Convert "Height" column to numeric values
df["Height"] = pd.to_numeric(df["Height"], errors="coerce")
```

```python
# Remove rows with NaN (not-a-number) values if necessary

df=df.dropna(subset=["Weight","Height"])
```

```python
# Perform linear regression
result=linregress(df["Weight"] , df["Height"])
```

```python
# print the result
print(result)
```
