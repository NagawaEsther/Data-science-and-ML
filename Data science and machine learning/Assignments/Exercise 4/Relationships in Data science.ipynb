```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (height and weight)
height = [150, 160, 165, 170, 175, 180]
weight = [50, 60, 65, 70, 75, 80]

# Plotting the data points
plt.plot(weight, height, "o")
plt.title("Relationship between height and weight in cohort 3 class")
plt.xlabel("Weight (kgs)")
plt.ylabel("Height (cm)")

# Calculate the line of best fit
# Perform linear regression using polyfit
coefficients = np.polyfit(weight, height, 1)
m, b = coefficients  # slope (m) and y-intercept (b)

# Generate the line of best fit using the slope and intercept
fit_line = m * np.array(weight) + b

# Plotting the line of best fit
plt.plot(weight, fit_line, color='red', label=f'Line of Best Fit: y = {m:.2f}x + {b:.2f}')
plt.legend()

# Display the plot
plt.show()
```
